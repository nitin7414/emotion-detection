```python
import snscrape.modules.twitter as stweet 
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
model_name= "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
def get_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", truncation="true")
    outputs = model(**input)
    scores = torch.softmax(outputs.logits, dim=1).detach().numpy()[0]
    labels = ["Negative", "Neutral", "Positive"]
    return labels[scores.argmax(), scores.max()]

limit = 500
tweet_list = []
query = "T20 World Cup 2024 OR cricket OR T20WC2024 lang:en since:2024-06-01 until:2025-12-10"

for i , tweet in  stweet.TwitterSearchScraper(query).get_items():
    if i >= 100:
        break
    label, confidence = await get_sentiment(tweet.rawContent)


tweet_list.append(
    tweet.user.username,
    tweet.date,
    tweet.rawContent,
    label, confidence
)

df = pd.DataFrame(tweet_list, columns=["Username", "Date", "Tweet","Sentiment", "Confidence"])
df.to_csv("sentiment.csv", index=False)
print("Data is saved to sentiment.csv ")
print(df.head())
```
